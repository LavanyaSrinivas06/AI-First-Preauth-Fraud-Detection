```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use("seaborn-v0_8-whitegrid")

DATA_PATH = Path("../data/raw/creditcard.csv")

assert DATA_PATH.exists(), "Dataset missing. Place it in data/raw/creditcard.csv"
df = pd.read_csv(DATA_PATH)
print(f"Loaded {df.shape[0]:,} rows × {df.shape[1]} columns")
df.head()
```

```python
df.info()
display(df.describe().T)
```

```python
##Check for missing values
missing = df.isna().sum()
missing[missing > 0]
```

```python
##Class distribution
fraud_counts = df["Class"].value_counts()
plt.figure(figsize=(5,4))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values, palette="viridis")
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

fraud_ratio = fraud_counts[1] / fraud_counts.sum() * 100
print(f"Fraudulent transactions: {fraud_ratio:.3f}%")
```

```python
###Amount Distribution
plt.figure(figsize=(8,4))
sns.histplot(df["Amount"], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount ($)")
plt.ylabel("Frequency")
plt.show()

print(df["Amount"].describe())
```

```python
###Time Distribution
plt.figure(figsize=(8,4))
sns.histplot(df["Time"], bins=50, kde=True)
plt.title("Transaction Time Distribution")
plt.xlabel("Seconds Since First Transaction")
plt.show()
```

```python
##Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (PCA Components)")
plt.show()
```

```python
##Fraud vs. Legit difference in Amount
legit = df[df["Class"] == 0]
fraud = df[df["Class"] == 1]

plt.figure(figsize=(8,4))
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Transaction Amount by Class")
plt.show()

print("Legit mean amount:", legit["Amount"].mean())
print("Fraud mean amount:", fraud["Amount"].mean())
```

```python
###feature distributions PCA components
pca_cols = [f"V{i}" for i in range(1, 29)]
plt.figure(figsize=(12,6))
sns.boxplot(data=df[pca_cols].iloc[:, :10])  # first 10 for readability
plt.xticks(rotation=45)
plt.title("Distribution of PCA Components (first 10)")
plt.show()
```
